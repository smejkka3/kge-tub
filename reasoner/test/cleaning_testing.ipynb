```python
import pandas as pd
```

```python
def clean_dataset(df, schema=False):
    df[0] = df[0].str.replace("<|>|:","",regex=True).str.split('/').str[-1]
    df[1] = df[1].str.replace("<|>|:","",regex=True).str.split('/').str[-1].str.split('#').str[-1]
    df[2] = df[2].str.replace("<|>|:","",regex=True).str.split('/').str[-1].str.split('#').str[-1]
    if schema:
        df = df[df[1].str.contains('type|subClassOf|subPropertyOf|range|domain')]
    df = df[df[0].notna()]
    df = df.dropna().reset_index(drop=True)
    return df
```

```python
path = '/home/karel/Documents/github/thesis/data/yago/yago-wd-schema.nt'

df_schema = pd.read_csv(path, sep='\t',header=None,
                         names=None,
                         dtype=str,
                         usecols=[0, 1, 2])
df_schema = df_schema.drop_duplicates()
df_schema
```

```python
df_schema_clean = clean_dataset(df_schema, schema=True)
```

```python
path = '/home/karel/Documents/github/thesis/data/yago/yago-wd-full-types.nt'

df_types = pd.read_csv(path, sep='\t',header=None,
                         names=None,
                         dtype=str,
                         usecols=[0, 1, 2])
df_types = df_types.drop_duplicates()
print(df_types.info())
df_types
```

```python
df_types_sample = df_types.sample(n=10000, random_state=1)
df_types_sample
```

```python
df_types_sample_clean = clean_dataset(df_types_sample)
```

```python
path = '/home/karel/Documents/github/thesis/data/yago/yago-wd-facts.nt'

df_facts = pd.read_csv(path, sep='\t',header=None,
                         names=None,
                         dtype=str,
                         usecols=[0, 1, 2])
df_facts = df_facts.drop_duplicates()
print(df_facts.info())
df_facts
```

```python
df_facts_sample = df_facts.sample(n=10000, random_state=1)
df_facts_sample
```

```python
df_facts_sample_clean = clean_dataset(df_facts_sample)
```

```python
dfs = [df_facts_sample_clean, df_types_sample, df_schema_clean]
resulting_df = pd.concat(dfs).reset_index(drop=True)
```
