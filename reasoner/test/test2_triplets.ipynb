```python
triple = Term('triple')
subclass = Term('subclass')
type_ = Term('type')
subproperty = Term('subproperty')
range_ = Term('range')
domain_ = Term('domain')

c = Term('c')
x = Term('x')
y = Term('y')
z = Term('z')
c1 = Term('c1')
c2 = Term('c2')
c3 = Term('c3')
r = Term('r')
r1 = Term('r1')
r2 = Term('r2')
r3 = Term('r3')

C = Var('C')
C3 = Var('C3')
C2 = Var('C2')
C1 = Var('C1')
X = Var('X')
Y = Var('Y')
Z = Var('Z')
R = Var('R')
R1 = Var('R1')
R2 = Var('R2')
R3 = Var('R3')
```

```python

# triple(C1, subclass, C3) << triple(C1, subclass, C2), triple(C2, subclass, C3)
# triple(X, type_, C2) << triple(C1, subclass, C2), triple(X, type_, C1)
# triple(X, type_, C) << triple(R, domain_, C), triple(X, R, Y)
# triple(Y, type_, C) << triple(R, range_, C), triple(X, R, Y)
# triple(R1, subproperty, R3) << triple(R1, subproperty, R2), triple(R2, subproperty, R3)
# triple(X, R2, Y) << triple(R1, subproperty, R2), triple(X, R1, Y)


rules = PrologString(f"""
{triple}({X}, {type_}, {C}) :- {triple}({R}, {domain_}, {C}), {triple}({X}, {R}, {Y}).
{triple}({Y}, {type_}, {C}) :- {triple}({R}, {range_}, {C}), {triple}({X}, {R}, {Y}).
{triple}({C1}, {subclass}, {C3}) :- {triple}({C1}, {subclass}, {C2}), {triple}({C2}, {subclass}, {C3}).
{triple}({X}, {type_}, {C2}) :- {triple}({C1}, {subclass}, {C2}), {triple}({X}, {type_}, {C1}).
{triple}({X}, {R2}, {Y}) :- {triple}({R1}, {subproperty}, {R2}), {triple}({X}, {R1}, {Y}).

query({triple}({C1}, {subclass}, {C3})).

query({triple}({X}, {type_}, {C2})).

query({triple}({X}, {type_}, {C})).

query({triple}({Y}, {type_}, {C})).

query({triple}({X}, {R2}, {Y})).
""")

get_evaluatable().create_from(rules).evaluate()
```

```python
################################################################################
## RDFS 7
################################################################################

triple(X, R2, Y) << triple(R1, subproperty, R2), triple(X, R1, Y)

rdfs7 = PrologString(f"""
{triple}({r1}, {subproperty}, {r2}).
{triple}({x}, {r1}, {y}).

{triple}({X}, {R2}, {Y}) :- {triple}({R1}, {subproperty}, {R2}), {triple}({X}, {R1}, {Y}).

query({triple}({x}, {r2}, {y})).
""")

get_evaluatable().create_from(rdfs7).evaluate()
```

```python
################################################################################
## RDFS 9
################################################################################
c1 = 'country'
c2 = 'administrativeArea'
x = 'united_States'

#<http://schema.org/Country>	<http://www.w3.org/2000/01/rdf-schema#subClassOf>	<http://schema.org/AdministrativeArea>	. (class)
#<http://yago-knowledge.org/resource/United_States>	<http://www.w3.org/1999/02/22-rdf-syntax-ns#type>	<http://schema.org/Country>	. (types)

triple(X, type_, C2) << triple(C1, subclass, C2), triple(X, type_, C1)

rdfs9 = PrologString(f"""
{triple}({c1}, {subclass}, {c2}).
{triple}({x}, {type_}, {c1}).

{triple}({X}, {subclass}, {C2}) :- {triple}({C1}, {subclass}, {C2}), {triple}({X}, {type_}, {C1}).

query({triple}({x}, {subclass}, {C2})).
""")

get_evaluatable().create_from(rdfs9).evaluate()
```
